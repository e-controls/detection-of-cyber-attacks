```python
import pandas as pd
import matplotlib.pyplot as plt

# Criar DataFrame com os dados fornecidos
data_legit = {
    'Algoritmo': ['Decision Tree', 'Random Forest', 'Gradient Boosting', 'Naive Bayes', 'Multilayer Perceptron', 'Sequential Neural Network'],
    'Precisão': [0.94, 0.94, 0.95, 0.61, 0.87, 0.87],
    'Recall': [0.98, 0.98, 0.96, 1.00, 0.99, 0.99]
}

data_malicious = {
    'Algoritmo': ['Decision Tree', 'Random Forest', 'Gradient Boosting', 'Naive Bayes', 'Multilayer Perceptron', 'Sequential Neural Network'],
    'Precisão': [0.98, 0.98, 0.96, 1.00, 0.98, 0.99],
    'Recall': [0.94, 0.94, 0.95, 0.35, 0.86, 0.86]
}

df_legit = pd.DataFrame(data_legit)
df_malicious = pd.DataFrame(data_malicious)

# Gráfico combinado de Precisão e Recall para tráfego legítimo e malicioso
plt.figure(figsize=(14, 7))

# Tráfego Legítimo
plt.plot(df_legit['Algoritmo'], df_legit['Precisão'], marker='o', linestyle='-', color='b', label='Precisão - Legítimo')
plt.plot(df_legit['Algoritmo'], df_legit['Recall'], marker='o', linestyle='-', color='r', label='Recall - Legítimo')

# Tráfego Malicioso
plt.plot(df_malicious['Algoritmo'], df_malicious['Precisão'], marker='o', linestyle='--', color='b', label='Precisão - Malicioso')
plt.plot(df_malicious['Algoritmo'], df_malicious['Recall'], marker='o', linestyle='--', color='r', label='Recall - Malicioso')

plt.title('Precisão e Recall de Diferentes Algoritmos para Tráfego Legítimo e Malicioso')
plt.xlabel('Algoritmo')
plt.ylabel('Métrica')
plt.xticks(rotation=45, ha='right', fontsize=12)  # Aumentar o tamanho da fonte para 12
plt.yticks(fontsize=10)
plt.legend()
plt.grid(True)
plt.show()
```
